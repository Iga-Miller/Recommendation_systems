```python
import pandas as pd
from pathlib import Path
```

```python
DATA_DIR = Path("./recsys/data/milion_songs")
PREPRO_WORK_DIR = DATA_DIR / "preprocessed"
```

```python
music_info_df = pd.read_csv(DATA_DIR / "music_info.csv").drop(columns=['spotify_preview_url', 'spotify_id'])
music_info_df
```

```python
music_info_df.info()
```

```python
music_info_df.to_pickle(PREPRO_WORK_DIR / "music_info_df.pkl")
```

```python
user_listening_history_df = pd.read_csv(DATA_DIR / "user_listening_history.csv").iloc[1:]
user_listening_history_df
```

```python
user_listening_history_df.info()
```

```python
user_listening_history_df.sort_values('playcount', ascending=True)
```

```python
def filter_songs(df: pd.DataFrame):
    new_df = df.copy()
    new_df = new_df[new_df['playcount'] > 20]
    return new_df

user_listening_history_df = filter_songs(user_listening_history_df).reset_index(drop=True)
user_listening_history_df
```

```python
user_listening_history_df.to_pickle(PREPRO_WORK_DIR / "user_listening_history_df.pkl")
```
