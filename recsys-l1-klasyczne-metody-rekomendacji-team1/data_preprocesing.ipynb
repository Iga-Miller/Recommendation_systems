```python
import pandas as pd
from pathlib import Path
```

```python
DATA_DIR = Path("../data/steam")
PREPRO_WORK_DIR = DATA_DIR / "preprocessed"
```

```python
users_df = pd.read_pickle(PREPRO_WORK_DIR / "users_df.pkl")
users_df
```

```python
users_df.info()
```

```python
# drop_duplicates work around for list/array columns
users_df = users_df[~users_df.astype({"items": str, "reviews": str}).duplicated()]
users_df.shape
```

```python
users_df['reviews'][0]
```

```python
def filter_users(df: pd.DataFrame):
    new_df = df.copy()
    new_df = new_df[new_df['reviews'].apply(lambda x : len(x) >=5)]
    new_df = new_df[new_df['items_count'] > 0]
    return new_df

users_df = filter_users(users_df).reset_index(drop=True)
users_df
```

```python
users_df['items'][0][:5]
```

```python
users_df['steam_id'].nunique()
```

```python
owned_games_df = users_df.explode("items", ignore_index=True)
owned_games_df
```

```python
owned_games_df["items"][0]
```

```python
games_temp_df = pd.DataFrame.from_records(owned_games_df["items"])
games_temp_df
```

```python
pivot_temp_df = pd.concat([owned_games_df, games_temp_df], axis=1)
pivot_temp_df
```

# Recommend score?

```python
pivot_temp_df["recommend_score"] = pivot_temp_df["playtime_forever"] + 3 * pivot_temp_df["playtime_2weeks"] + 100 * pivot_temp_df["items_count"]
```

```python
pivot_users_df = pd.pivot_table(pivot_temp_df, index="steam_id", columns="item_id", values="recommend_score")
pivot_users_df = pivot_users_df.fillna(-1)
pivot_users_df
```

```python
# assert that `steam_id` and `item_id` are unique
assert pivot_users_df.index.nunique() == pivot_users_df.shape[0]
assert pivot_users_df.keys().nunique() == pivot_users_df.shape[1]
```

```python
pivot_users_df.to_pickle(PREPRO_WORK_DIR / "pivot_users_games_recscore_df.pkl")
```
