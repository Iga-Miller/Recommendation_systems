```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
df = pd.read_pickle('Metadata.pkl')
df
```

```python
cosine_similarity_text = cosine_similarity(df['description'].tolist())
cosine_similarity_image = cosine_similarity(df['image_emb'].tolist())
cosine_similarity = (cosine_similarity_text + cosine_similarity_image) / 2
```

```python
cosine_similarity.shape
```

```python
df = df.reset_index()
```

```python
def get_recommendations(idx, cosine_sim, df, num_recommend=10):
    similarity_scores = cosine_sim[idx]

    top_similar_indices = np.argsort(similarity_scores)[::-1][1:num_recommend+1]

    return df.iloc[top_similar_indices]
```

```python
get_recommendations(0, cosine_similarity, df, num_recommend=10)
```
